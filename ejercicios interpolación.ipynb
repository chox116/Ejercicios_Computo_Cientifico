1. Haz tu propia función llamada lagrange(n,x,y), que reciba los nodos o puntos para
interpolar x,y y el grado del polinomio deseado n.

```python
from pylab import *
from IPython.display import Image
from random import *
from math import * 
from numpy import *
```

```python
def Lagrange_n(x, X, Y):                #Lagrande para polinomios de grado n
    f=0                                 #definimos f=0 para la suma 
    L=1                                 #neutro multiplicativo
    for i in range(len(X)):             #inician los contadores dependiendo del grado del polinomio deseado(len(X))
        for j in range(len(X)):
            if(i!=j):                   #restringes el productorio para evitar la división entre 0 
                L*=(x-X[j])/(X[i]-X[j]) #L es el productorio y se multiplica
        f+=L*Y[i]                       #esta es la suma de los polinomios de Lagrange 
        L=1                             #regresas a definir L=1 para volver a empezar la interpolación con alguna otra función 
    return f
```

2. Sea $f(x) = 2 sin(πx/6)$ (con el ángulo medido en radianes).

a)Use el polinomio de Lagrangre P2 con nodos x0 = 0, x1 = 1 y x2 = 3 para aproximar
f(2) y f(2.4),

b)Use el polinomio de Lagrangre P3 con nodos x0 = 0, x1 = 1, x2 = 3 y x3 = 5 para
aproximar f(2) y f(2.4)

c)Use el polinomio de Lagrangre P2 con nodos x0 = 0, x1 = 1 y x2 = 3 para aproximar
f(4) y f(3.5),

d)Use el polinomio de Lagrangre P3 con nodos x0 = 0, x1 = 1, x2 = 3 y x3 = 5 para
aproximar f(4) y f(3.5),

```python
def cua_interp(x, p0, p1,p2): #Lagrange polinomios grado 2 para comparar con la funcion de grado n que se programo arriba
    x0, y0 = p0
    x1, y1 = p1
    x2, y2 = p2
    return y0*((x-x1)*(x-x2))/((x0-x1)*(x0-x2))+y1*((x-x0)*(x-x2))/((x1-x0)*(x1-x2))+y2*((x-x0)*(x-x1))/((x2-x0)*(x2-x1))

def g(x):
    return 2*sin((pi*x)/6) #definimos la función 
```

```python
# a) f(2) y f(2.4), y c) f(4) y f(3.5), polinomios grado 2
#Nodos y valores de la función 
X=array([0,1,3])
Y=array(g(X))
p=list(zip(X,Y))


print("El valor real de la función en 2 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(2.), Lagrange_n(2., X, Y), cua_interp(2., p[0], p[1], p[2]))) 
print("El valor real de la función en 2.4 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(2.4), Lagrange_n(2.4, X, Y), cua_interp(2.4, p[0], p[1], p[2])))
print("\n El valor real de la función en 3.5 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(3.5), Lagrange_n(3.5, X, Y), cua_interp(3.5, p[0], p[1], p[2])))
print("El valor real de la función en 4 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(4), Lagrange_n(4, X, Y), cua_interp(4, p[0], p[1], p[2])))

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20, 9))
xr = np.linspace(0, 3)
ax1.plot(xr, g(xr), 0.1, label=r"$y = \frac{2sin(\pi x)}{6} $")
ax1.plot(xr, Lagrange_n(xr, X, Y), 'r:', label=r"$y = Lagrange_3(x)$")
ax1.legend()

xr = np.linspace(0, 3)
ax2.plot(xr, g(xr), label=r"$y = \frac{2sin(\pi x)}{6} $")
ax2.plot(xr, cua_interp(xr, p[0], p[1],p[2]), ":", label=r"$y = cua_interp(x)$")
ax2.legend()
```

```python
#b)f(2) y f(2.4) y d) f(4) y f(3.5).

def tercer_interp(x,p0,p1,p2,p3):
    x0, y0 = p0
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    return y0*((x-x1)*(x-x2)*(x-x3))/((x0-x1)*(x0-x2)*(x0-x3))+y1*((x-x0)*(x-x2)*(x-x3))/((x1-x0)*(x1-x2)*(x1-x3))+y2*((x-x0)*(x-x1)*(x-x3))/((x2-x0)*(x2-x1)*(x2-x3))+ y3*((x-x0)*(x-x1)*(x-x2))/((x3-x0)*(x3-x1)*(x3-x2))

#nodos y valores de la función 
X1=array([0, 1, 3, 5])
Y1=array(g(X1))
p1=list(zip(X1,Y1))

print("El valor real de la función en 2 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 3 es %s \n" %(g(2.), Lagrange_n(2., X1, Y1), tercer_interp(2., p1[0], p1[1], p1[2], p1[3]))) 
print("El valor real de la función en 2.4 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(2.4), Lagrange_n(2.4, X1, Y1), tercer_interp(2.4, p1[0], p1[1], p1[2], p1[3])))
print("\n El valor real de la función en 3.5 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(3.5), Lagrange_n(3.5, X1, Y1), tercer_interp(3.5, p[0], p[1], p[2], p1[3])))
print("El valor real de la función en 4 es %s y la aproximación con Lagrange_n que se programó es %s y con la función de grado 2 es %s \n" %(g(4.), Lagrange_n(4, X1, Y1), tercer_interp(4, p1[0], p1[1], p1[2], p1[3])))

fig, (ax3, ax4) = plt.subplots(ncols=2, figsize=(20, 9))
xr = np.linspace(0, 5)
ax3.plot(xr, g(xr), 0.1, label=r"$y = \frac{2sin(\pi x)}{6} $")
ax3.plot(xr, Lagrange_n(xr, X1, Y1), 'r:', label=r"$y = Lagrange_3(x)$")
ax3.legend()

xr = np.linspace(0, 5)
ax4.plot(xr, g(xr), label=r"$y = \frac{2sin(\pi x)}{6} $")
ax4.plot(xr, tercer_interp(xr, p1[0], p1[1],p1[2], p1[3]), ":", label=r"$y = tercer_interp(x)$")
ax4.legend()
```
